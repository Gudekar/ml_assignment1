```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv('test.csv')
data
```

```python
data.info()
```

```python
plt.scatter(data.x,data.y,color='red')
```

```python
#seperate input
inp=data[['x']]
inp[:5]
```

```python
#seperate output
out= data['y']
out[:5]
```

```python
#training phase
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model
```

```python
model.fit(inp,out)
```

```python
model.score(inp,out)*100
```

```python
#testing phase
model.predict([[72]])
```

```python
#m
model.coef_
```

```python
#c
model.intercept_
```

```python
model.predict([[17]])
```

```python
#create test input data
test=data.sample(50)
test
```

```python
test_in=test[['x']]
test_in
test_out=test[['y']]
test_out
```

```python
test_pred=model.predict(test_in)
test_pred
```

```python
# plot training and testing result
plt.scatter(data.x,data.y,color='red',label='Data points')
plt.plot(test_in,test_pred,'k--',label='best fit line')
plt.legend()
```
